In the current plot the user can compare prices withing lego sets and understand how much the price of the set can change over time.

```python
# Visualisation NUMBER 7
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Category20
import pandas as pd

# Convert 'Date' column to datetime if it's not already
dft['Date'] = pd.to_datetime(dft['Date'])

# Sort the DataFrame by 'Date'
dft = dft.sort_values('Date')

# Filter the DataFrame to include only rows where the subtheme is "Ultimate Collector Series"
ultimate_collector_df = df[df['subtheme'] == 'Ultimate Collector Series']

# Create a figure
p = figure(x_axis_type='datetime', title="Ultimate Collector Series - Price Changes", width=1000, height=600)

# Loop through each item in the Ultimate Collector Series
for i, item_name in enumerate(ultimate_collector_df['name'].unique()):
    item_data = dft[dft['name'] == item_name]
    
    # Create a ColumnDataSource for each item
    source = ColumnDataSource(data=dict(
        Date=item_data['Date'],
        PriceMonthPLN=item_data['PriceMonthPLN'],
        Name=[item_name] * len(item_data),
    ))
    
    # Plot the line
    p.line('Date', 'PriceMonthPLN', source=source, legend_label=item_name, line_color=Category20[20][i%20], visible=False)
    
    # Add circle markers
    p.circle('Date', 'PriceMonthPLN', source=source, size=6, color=Category20[20][i%20], legend_label=item_name, visible=False)

# Add hover tool
hover = HoverTool(tooltips=[("Name", "@Name"), ("Price", "@PriceMonthPLN{0.00}")])
p.add_tools(hover)

# Customize plot
p.xaxis.axis_label = "Date"
p.yaxis.axis_label = "Price (Euro)"
p.legend.location = "top_right"
p.legend.click_policy = "hide"  # Click on legend to hide/show lines

# Move legend outside of the plot area
p.legend.orientation = "vertical"
p.legend.border_line_alpha = 1
p.legend.background_fill_alpha = 1

# Adjust the size of the legend
p.legend.label_text_font_size = "10pt"

# Show the plot
show(p)
```

In this visualization we decide to use the available dates from each coutry so the user could understan the time window he had to buy a new lego set. It is also helful to see how much time an itel is available in the market before it moves on the the second had market.

```python
# Visualisation NUMBER 8
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6

# Specify the output file
output_file("timewindow.html")

# Filter the DataFrame to include only rows where the subtheme is "Ultimate Collector Series"
ultimate_collector_df = df[df['subtheme'] == 'Ultimate Collector Series']

# Group the DataFrame by 'name' and get the first and last available dates for each product in each country
grouped_df = ultimate_collector_df.groupby('name').agg({
    'US_dateFirstAvailable': 'min', 'US_dateLastAvailable': 'max',
    'UK_dateFirstAvailable': 'min', 'UK_dateLastAvailable': 'max',
    'CA_dateFirstAvailable': 'min', 'CA_dateLastAvailable': 'max',
    'DE_dateFirstAvailable': 'min', 'DE_dateLastAvailable': 'max'
})

# Reset the index
grouped_df = grouped_df.reset_index()

# Convert the DataFrame to a ColumnDataSource
source = ColumnDataSource(grouped_df)

# Create a figure
p = figure(x_range=(grouped_df[['US_dateFirstAvailable', 'UK_dateFirstAvailable', 'CA_dateFirstAvailable', 'DE_dateFirstAvailable']].min().min(), 
                    grouped_df[['US_dateLastAvailable', 'UK_dateLastAvailable', 'CA_dateLastAvailable', 'DE_dateLastAvailable']].max().max()), 
           y_range=grouped_df['name'], height=600, width=800, 
           title="Time window for Ultimate Collector Series products in Each Country",
           toolbar_location=None, tools="")

# Use hbar to create the colored rectangles for each country
p.hbar(y=dodge('name', -0.3, range=p.y_range), left='US_dateFirstAvailable', right='US_dateLastAvailable', height=0.2, color=Spectral6[0], source=source, legend_label='US')
p.hbar(y=dodge('name', -0.1, range=p.y_range), left='UK_dateFirstAvailable', right='UK_dateLastAvailable', height=0.2, color=Spectral6[1], source=source, legend_label='UK')
p.hbar(y=dodge('name', 0.1, range=p.y_range), left='CA_dateFirstAvailable', right='CA_dateLastAvailable', height=0.2, color=Spectral6[2], source=source, legend_label='CA')
p.hbar(y=dodge('name', 0.3, range=p.y_range), left='DE_dateFirstAvailable', right='DE_dateLastAvailable', height=0.2, color=Spectral6[3], source=source, legend_label='DE')

# Customize the plot
p.x_range.start = grouped_df[['US_dateFirstAvailable', 'UK_dateFirstAvailable', 'CA_dateFirstAvailable', 'DE_dateFirstAvailable']].min().min()
p.y_range.range_padding = 0.1
p.ygrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_right"
p.legend.orientation = "vertical"

# Show the plot
show(p)
```
